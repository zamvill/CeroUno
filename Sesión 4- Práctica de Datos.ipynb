# 1. Introducción a un set de datos reales Iniciativa #3de3


Bajar el set de datos la iniciativa #3de3

Descargar información de https://www.3de3.mx/datos_abiertos

```python
import requests
import pandas as pd
```

```python
import requests
import time

#Se bajaron registros solo con los siguientes filtros de entidades y partidos

df=[]
paginas=['1','2','3','4']
generos=['F','M']
entidades=['Nuevo León', 'Coahuila','San Luis','Tamaulipas','Sinaloa','Queretaro','Chihuahua','Sonora','Baja California Sur','Baja California Norte','Colima','Aguscalientes','Guerrero',]
partidos=['PRI', 'PAN', 'PT', 'PVEM', 'PRD', 'MC', 'Nueva Alianza', 'Morena', 'PH', 'PES', 'Otro', 'Independiente']
for pagina in paginas:
    for genero in generos:
        for partido in partidos:
            for entidad in entidades:
                print(str(pagina)+ ' ' + str(genero) + ' '+ str(partido)+' '+str(entidad))

                headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
                url = 'https://3de3.mx/declarantes.json?page='+pagina+'&by_gender='+genero+'&by_political_party='+partido+'&by_federal_entity='+entidad
                declarantes_json = requests.get(url, headers=headers)

                #declarantes_json = requests.get("https://3de3.mx/declarantes.json?page="+pagina+'&by_gender='+genero+'&by_political_party='+partido)
                print(declarantes_json)
                declarantes = pd.DataFrame(declarantes_json.json())
                declarantes['Genero'] =genero
                declarantes['Partido'] =partido
                declarantes['Entidad'] =entidad
                df.append(declarantes)
                print(type(df))
                print(type(df[0]))
                time.sleep(1)
        
df=pd.concat(df).reset_index()
df.head()
    
```
