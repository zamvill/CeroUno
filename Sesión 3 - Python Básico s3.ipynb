```python
#abre un archivo de lectura y regresa el contenido en una variable
#recibe como parametro el nombre del archivo
def readfile(filen):
    # Se lee archivo y se guarda en una variable
    archivo = open(filen)
    contenido=archivo.read()
    #print(o)
    archivo.close()
    return contenido

#elimina signos de puntuacion en un string
#recibe como parametro un string
def deletesigns(s):
    #se declaran signos de puntacion para ser eliminados despues en caso de exisir en el texto
    punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    #longitud del string de signos de puntuacion
    lp=len(punctuation)
    
    #primer for para recorrer el string de signos de puntuacion
    for i in range(0,lp-1):
            #si el signo se encuentra en el string de 5mil, se elimina signo
            if s.find(punctuation[i]) > 0:
                s=s.replace(punctuation[i], '')
                
    return s

#encuentra signos de puntacion y regresa en un string los que encontro
#parametro string
def findsignos(s):
    #se declaran signos de puntacion para ser eliminados despues en caso de exisir en el texto
    punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    #longitud del string de signos de puntuacion
    lp=len(punctuation)
    ###proceso de eliminar signos de puntuacion
    #Variable para guardar signos que se van a eliminar
    removedsign=""
    #primer for para recorrer el string de signos de puntuacion
    for i in range(0,lp-1):
            #si el signo se encuentra en el string, se elimina signo
            if s.find(punctuation[i]) > 0:
                #se guarda el signo eliminado
                removedsign= removedsign +punctuation[i]
    return removedsign
    
#parametro un string y regresa una lista de palabras    
def stringtolista(s):
    #string original en lista
    lista_palabras=s.split()
    return lista_palabras

#parametro una lista de palabras y regresa el numero de elementos en la lista
def nopalabras(l):
    palabras=len(l) 
    return palabras

#parametro una lista de palabras y elimina los elementos repetidos
def palabras_unicas(lista):
    p=set(lista)
    return p

#parametro lista de palabras y regresa palabras unicas sin repetir
def nupalabras_unicas(lista):
    n=len(set(lista))
    return n

#parametros lista de palabras y palabra a buscar en la lista, regresa el numero de veces encontrada esa palabra
#en la lista
def findwordveces(lista,word):
    n=lista.count(word)
    return n
#parametro un string y lo regresa en mayusculas
def changetoupper(s):
    mi_stringUpper=s.upper()
    return mi_stringUpper

#parametro recibe lista de palabras y regresa un diccionario {palabra:Numero de veces encontrada}
def glosario(s):
    ##############################################
    # Crear Glosario de palabras en un diccionario
    ###############################################
    glosario={}
    for i in palabras_unicas(s):
        glosario.update({i:lista_palabras.count(i)})
    ## Se imprime el diccionarion
    a=1
    for (key, value) in sorted(glosario.items()) :
            print(str(a), key , " :: ", value )
            a+=1
    

#El contenido del archivo de texto se asigna a la variable
texto=readfile("8800.txt")

print("¿Cuántas palabras existen en el archivo?  ==> "+str(nopalabras(stringtolista(texto)))+" palabras")
print("¿Cuántas palabras únicas existen en el mismo rango?  ==> "+str(nupalabras_unicas(stringtolista(texto)))+" palabras unicas")
print("¿Importan las mayúsculas y minúsculas? ==> Si")
print("-----------EN MAYUSCULAS--------------")
print("¿Cuántas palabras únicas en mayusculas existen en el mismo rango?  ==>"+str(nupalabras_unicas(stringtolista(changetoupper(texto))))+" palabras unicas")

print("¿Cuántas veces aparece la palabra the en ese rango?  ==> " +str(findwordveces(stringtolista(texto),"the"))+" veces")
print("¿Puedo saber cuantas veces existe una palabra en el texto cargado? ==> Si")
print("¿Qué pasa con los signos de puntuación, es decir, 'me', 'me;', 'me.', 'me?' y 'me,' son iguales? ==> No")
print("¿Cuáles existen? ==> " + findsignos(texto))
print("Crea un diccionario donde asignes el número de apariciones a la palabra existente")
print("Glosario")

#Glosario sin eliminar signos de puntacion
#glosario(stringtolista(texto))

#Glosario sin singnos de puntacion
glosario(stringtolista(deletesigns(texto)))
```
